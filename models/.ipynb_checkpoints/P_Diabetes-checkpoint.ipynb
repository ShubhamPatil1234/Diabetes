# Domain Knowledge

```python
"""We have an dataset having following columns with given information related to Diabetes Disease"""

Features of dataset are as follows:
    
1. Glucose: Plasma glucose concentration a 2 hours in an oral glucose tolerance test
2. BloodPressure: Diastolic blood pressure (mm Hg)
3. SkinThickness: Triceps skin fold thickness (mm)
4. Insulin: 2-Hour serum insulin (mu U/ml)
5. BMI: Body mass index (weight in kg/(height in m)^2)
6. DiabetesPedigreeFunction: Diabetes pedigree function
7. Age: Age (years)
8. Outcome: Class variable (0 or 1)
```

```python
x = df.drop('Outcome', axis=1)
y = df['Outcome']  # target column

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=1, stratify=y)
```

```python
# Testing Data
plot_confusion_matrix(dibetes_model, x_test, y_test)
```

# decision tree prunning

```python
output = dt_clf.cost_complexity_pruning_path(x_train, y_train)
ccp_alpha_list = output['ccp_alphas']
ccp_alpha_list
```

```python
train_accuracy_list = []
test_accuracy_list = []
for i in ccp_alpha_list:
    dt_clf_pp_model = DecisionTreeClassifier(ccp_alpha= i,random_state=10)
    dt_clf_pp_model.fit(x_train, y_train)

    training_accuracy = dt_clf_pp_model.score(x_train, y_train)
    train_accuracy_list.append(training_accuracy)

    testing_Accuracy = dt_clf_pp_model.score(x_test, y_test)
    test_accuracy_list.append(testing_Accuracy)
```

```python
fig, ax = plt.subplots()
ax.plot(ccp_alpha_list, train_accuracy_list, label = "Training Data Accuracy")
ax.plot(ccp_alpha_list, test_accuracy_list, label = "Testing Data Accuracy")
ax.legend()
```

```python
index_ccp = test_accuracy_list.index(max(test_accuracy_list))
index_ccp
```

```python
best_ccp = ccp_alpha_list[index_ccp]

dt_clf_model = DecisionTreeClassifier(ccp_alpha= best_ccp,random_state=6)
dt_clf_model.fit(x_train, y_train)
model_details.append("decision tree prunning model")
```

```python
# Training data evaluation

y_pred_train = dt_clf_model.predict(x_train)

cnf_matrix = confusion_matrix(y_train, y_pred_train)
print("Confusion Matrix :\n", cnf_matrix)

print("*"*80)
accuracy = accuracy_score(y_train, y_pred_train)
print("Accuracy Score with hyperparameter tuning on training:", accuracy)
print("*"*80)

clf_report = classification_report(y_train, y_pred_train)
print("Classification report :\n", clf_report)
Training_accuracy.append(accuracy)

# testing data evaluation
y_pred = dt_clf_model.predict(x_test)

cnf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix :\n", cnf_matrix)

print("*"*80)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score with hyperparameter tuning on testing:", accuracy)
print("*"*80)

clf_report = classification_report(y_test, y_pred)
print("Classification report :\n", clf_report)
Testing_accuracy.append(accuracy)
```
